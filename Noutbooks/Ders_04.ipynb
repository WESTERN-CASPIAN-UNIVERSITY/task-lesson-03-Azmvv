```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Məlumatları hazırlayırıq (7 günlük addım sayıları)
gunler = ['B.E', 'Ç.A', 'Çərs', 'C.A', 'Cümə', 'Şənbə', 'Bazar']
addimlar = np.array([5200, 7100, 4800, 10500, 8900, 15200, 12300])

# 2. Statistik hesablamalar (NumPy ilə)
cem_addim = np.sum(addimlar)        # Həftəlik cəmi addım
orta_hesab = np.mean(addimlar)      # Günlük orta addım
maksimum = np.max(addimlar)         # Ən aktiv günün addımı
minimum = np.min(addimlar)          # Ən tənbəl günün addımı

# 3. Nəticələri terminalda göstəririk
print("--- Həftəlik Aktivlik Hesabatı ---")
print(f"Ümumi atılan addım: {cem_addim}")
print(f"Günlük orta addım: {orta_hesab:.0f}")
print(f"Ən yüksək nəticə: {maksimum}")
print(f"Ən aşağı nəticə: {minimum}")
print("----------------------------------")

# 4. Vizuallaşdırma (Matplotlib ilə)
plt.figure(figsize=(10, 6))

# Addımların dəyişməsini göstərən xətt qrafiki
plt.plot(gunler, addimlar, marker='o', color='teal', linestyle='-', linewidth=2, label='Günlük Addım')

# Orta hesabı göstərən üfüqi xətt
plt.axhline(y=orta_hesab, color='red', linestyle='--', label=f'Orta ({orta_hesab:.0f})')

# Qrafikin tənzimlənməsi
plt.title('Həftəlik Addım Sayı Analizi', fontsize=14)
plt.xlabel('Həftənin Günləri')
plt.ylabel('Addım Sayı')
plt.grid(True, alpha=0.3)
plt.legend()

# Qrafiki ekranda göstəririk
plt.show()
```
